```python
soup.find_all('div', class_ = 'match-block__teams-wrapper match-block__teams-wrapper--desktop col-8')
```

```python
match_divs = soup.find_all('div', class_='match-block__teams-wrapper match-block__teams-wrapper--desktop col-8')

# Loop through each match div
for match in match_divs:
    team_info = match.find_all('div', class_='match-block__team')
    
    for info in team_info:
        team_name = info.find('span', class_='match-block__team-name').text.strip()
        score = info.find('span', class_='match-block__score').text.strip()
        overs_played = info.find('span', class_='match-block__score--overs').text.strip()
        
        print(f'Team: {team_name}')
        print(f'Score: {score}')
        print(f'Overs Played: {overs_played}')
        print('------------------------------------------------------------')
```

```python

match_divs = soup.find_all('div', class_='match-block__teams-wrapper match-block__teams-wrapper--desktop col-8')

teams_list = []  # List to store team names

# Extract team information from each match div
for match in match_divs:
    team_info = match.find_all('div', class_='match-block__team')
    
    for info in team_info:
        team_name = info.find('span', class_='match-block__team-name').text.strip()
        teams_list.append(team_name)  # Append team names to the list

# Pair the teams
paired_teams = [(teams_list[i], teams_list[i + 1]) for i in range(0, len(teams_list), 2)]

# Print pairs in the format "1&2, 3&4, etc."
for index, Match in enumerate(paired_teams, start=1):
    team_pair = f"{Match[0]} & {Match[1]}"
    print(f"Match {index}: {Match}")
```
